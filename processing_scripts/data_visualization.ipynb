```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("ticks")
```

```python
date = '20190716'
percentage = '10'
path = f'../raw_data/director_mentions/slice_data/{percentage}'
```

## 1
кривые (x, y) = (date, mentions) для мужчин и женщин на одном графике для каждого из трех издания на основе данных 

### Deadline mentions per month

```python
female_deadline_month = pd.read_csv(f'{path}/female_mentions_per_month_deadline_{date}.txt', sep = ' ', header = None, names = ['year', 'month', 'count'])
female_deadline_month['date'] = female_deadline_month.apply(lambda x: datetime.date(year=int(x[0]), month=int(x[1]), day=1), axis =1)
female_deadline_month = female_deadline_month.sort_values(by='date')
female_deadline_month['date'] = female_deadline_month['date'].apply(lambda x: str(x.year)+'-'+str(x.month))
#female_deadline_month['relative'] = female_deadline_month['count'].apply(lambda x: x/51090)
```

```python
male_deadline_month = pd.read_csv(f'{path}/male_mentions_per_month_deadline_{date}.txt', sep = ' ', header = None, names = ['year', 'month', 'count'])
male_deadline_month['date'] = male_deadline_month.apply(lambda x: datetime.date(year=int(x[0]), month=int(x[1]), day=1), axis =1)
male_deadline_month = male_deadline_month.sort_values(by='date')
male_deadline_month['date'] = male_deadline_month['date'].apply(lambda x: str(x.year)+'-'+str(x.month))
#male_deadline_month['relative'] = male_deadline_month['count'].apply(lambda x: x/51090)
```

```python
deadline_month = male_deadline_month.merge(female_deadline_month, on = 'date', suffixes = ('_male', '_female'))
deadline_month['total'] = deadline_month.apply(lambda x: x[2] + x[-1], axis = 1)
deadline_month['relative_male'] = deadline_month.apply(lambda x: x[2]/x[-1], axis = 1)
deadline_month['relative_female'] = deadline_month.apply(lambda x: x[6]/x[-2], axis = 1)
```

### Holywood reporter mentions per month

```python
female_hwr_month = pd.read_csv(f'{path}/female_mentions_per_month_hollywood_reporter_{date}.txt', sep=' ', header = None, names = ['year', 'month', 'count'])
female_hwr_month['date'] = female_hwr_month.apply(lambda x: datetime.date(year=int(x[0]), month=int(x[1]), day=1), axis =1)
female_hwr_month = female_hwr_month.sort_values(by='date')
female_hwr_month['date'] = female_hwr_month['date'].apply(lambda x: str(x.year)+'-'+str(x.month))
#female_hwr_month['relative'] = female_hwr_month['count'].apply(lambda x: x/101880)
```

```python
male_hwr_month = pd.read_csv(f'{path}/male_mentions_per_month_hollywood_reporter_{date}.txt', sep=' ', header = None, names = ['year', 'month', 'count'])
male_hwr_month['date'] = male_hwr_month.apply(lambda x: datetime.date(year=int(x[0]), month=int(x[1]), day=1), axis =1)
male_hwr_month = male_hwr_month.sort_values(by='date')
male_hwr_month['date'] = male_hwr_month['date'].apply(lambda x: str(x.year)+'-'+str(x.month))
#male_hwr_month['relative'] = male_hwr_month['count'].apply(lambda x: x/101880)
```

```python
hwr_month = male_hwr_month.merge(female_hwr_month, on = 'date', suffixes = ('_male', '_female'))
hwr_month['total'] = hwr_month.apply(lambda x: x[2] + x[-1], axis = 1)
hwr_month['relative_male'] = hwr_month.apply(lambda x: x[2]/x[-1], axis = 1)
hwr_month['relative_female'] = hwr_month.apply(lambda x: x[6]/x[-2], axis = 1)
```

### Variety mentions per month

```python
female_variety_month = pd.read_csv(f'{path}/female_mentions_per_month_variety_{date}.txt', sep=' ', header = None, names = ['year', 'month', 'count'])
female_variety_month['date'] = female_variety_month.apply(lambda x: datetime.date(year=int(x[0]), month=int(x[1]), day=1), axis =1)
female_variety_month = female_variety_month.sort_values(by='date')
female_variety_month['date'] = female_variety_month['date'].apply(lambda x: str(x.year)+'-'+str(x.month))
#female_variety_month['relative'] = female_variety_month['count'].apply(lambda x: x/187228)
```

```python
male_variety_month = pd.read_csv(f'{path}/male_mentions_per_month_variety_{date}.txt', sep=' ', header = None, names = ['year', 'month', 'count'])
male_variety_month['date'] = male_variety_month.apply(lambda x: datetime.date(year=int(x[0]), month=int(x[1]), day=1), axis =1)
male_variety_month = male_variety_month.sort_values(by='date')
male_variety_month['date'] = male_variety_month['date'].apply(lambda x: str(x.year)+'-'+str(x.month))
#male_variety_month['relative'] = male_variety_month['count'].apply(lambda x: x/187228)
```

```python
variety_month = male_variety_month.merge(female_variety_month, on = 'date', suffixes = ('_male', '_female'))
variety_month['total'] = variety_month.apply(lambda x: x[2] + x[-1], axis = 1)
variety_month['relative_male'] = variety_month.apply(lambda x: x[2]/x[-1], axis = 1)
variety_month['relative_female'] = variety_month.apply(lambda x: x[6]/x[-2], axis = 1)
```

```python
variety_month.columns
```

### Plot mentions

```python
plt.figure(figsize=(22,10))

plt.title('Mentions of female and male directors per month')

plt.subplot(3,1,1)
sns.lineplot(x='date', y='relative_male', data = deadline_month, color = 'green')
sns.lineplot(x='date', y='relative_female', data = deadline_month, color = 'purple')
plt.title('Deadline per month')
plt.legend(('male', 'female'))
plt.xticks(range(len(deadline_month['date'].tolist())), deadline_month['date'].tolist())
plt.xticks(rotation=65)

plt.subplot(3,1,2)
sns.lineplot(x='date', y='relative_male', data = hwr_month, color = 'green')
sns.lineplot(x='date', y='relative_female', data = hwr_month, color = 'purple')
plt.title('Holywood Reporter per month')
plt.legend(('male', 'female'))
plt.xticks(rotation=65)
plt.xticks(range(len(hwr_month['date'].tolist())), hwr_month['date'].tolist())

plt.subplot(3,1,3)
sns.lineplot(x='date', y='relative_male', data = variety_month, color = 'green')
sns.lineplot(x='date', y='relative_female', data = variety_month, color = 'purple')
plt.title('Variety per month')
plt.legend(('male', 'female'))
plt.xticks(rotation=65)
plt.xticks(range(len(variety_month['date'].tolist())), variety_month['date'].tolist())

plt.tight_layout()
sns.despine()
plt.savefig(f'mentions_per_month_relative_{percentage}.png', dpi=300, format='png')
plt.show()
```

### Mentions by year

```python
deadline_year = deadline_month.groupby(['year_male'], as_index=False).sum()
hwr_year = hwr_month.groupby(['year_male'], as_index=False).sum()[2:]
variety_year = variety_month.groupby(['year_male'], as_index=False).sum()[2:]
```

```python
deadline_year['fem_year_relative'] = deadline_year.apply(lambda x: x[5]/x[6], axis = 1)
deadline_year['male_year_relative'] = deadline_year.apply(lambda x: x[2]/x[6], axis = 1)
```

```python
variety_year['fem_year_relative'] = variety_year.apply(lambda x: x[5]/x[6], axis = 1)
variety_year['male_year_relative'] = variety_year.apply(lambda x: x[2]/x[6], axis = 1)
```

```python
hwr_year['fem_year_relative'] = hwr_year.apply(lambda x: x[5]/x[6], axis = 1)
hwr_year['male_year_relative'] = hwr_year.apply(lambda x: x[2]/x[6], axis = 1)
```

```python
plt.figure(figsize=(5,10))

plt.subplot(3,1,1)
sns.barplot(x='year_male', y='male_year_relative', data = deadline_year.groupby(['year_male'], as_index=False).sum(), color = '#00BF85')
sns.barplot(x='year_male', y='fem_year_relative', data = deadline_year.groupby(['year_male'], as_index=False).sum(), color = 'purple')
plt.title('Deadline per month')
#plt.legend(('male', 'female'))
plt.xticks(rotation=65)

plt.subplot(3,1,2)
sns.barplot(x='year_male', y='male_year_relative', data = hwr_year.groupby(['year_male'], as_index=False).sum(), color = '#00BF85')
sns.barplot(x='year_male', y='fem_year_relative', data = hwr_year.groupby(['year_male'], as_index=False).sum(), color = 'purple')
plt.title('Holywood Reporter per month')
#plt.legend(('male', 'female'))
plt.xticks(rotation=65)

plt.subplot(3,1,3)
sns.barplot(x='year_male', y='male_year_relative', data = variety_year.groupby(['year_male'], as_index=False).sum(), color = '#00BF85')
sns.barplot(x='year_male', y='fem_year_relative', data = variety_year.groupby(['year_male'], as_index=False).sum(), color = 'purple')
plt.title('Variety per month')
#plt.legend(('male', 'female'))
plt.xticks(rotation=65)

plt.tight_layout()
sns.despine()
plt.savefig(f'mentions_year_barchart_relative_{percentage}.png', dpi=300, format='png')
plt.show()
```
